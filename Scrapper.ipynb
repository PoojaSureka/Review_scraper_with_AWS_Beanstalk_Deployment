```python
final  = pd.DataFrame()
for j in range(1,4):
    url = "https://www.flipkart.com/search?q=iphone11&page={}".format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text    
    
    soup = BeautifulSoup(webpage, "lxml")
    
    company = soup.find_all("div", class_ = "_2kHMtA")
    
    name = []
    rating = []
    price = []
   
    for i in company:

      try:
        name.append(i.find("div", {"class": "_4rR01T"}).text.strip())           
      except:
        name.append(np.nan)

      try:
        rating.append(i.find_all("span", {"class": "_2_R_DZ"})[0].text.strip())
      except:
        rating.append(np.nan)
      
      try:
        price.append(i.find("div", {"class": "_30jeq3 _1_WHN1"}).text)
      except:
        price.append(np.nan)

    d = {"name":name, "rating":rating,"price":price}

    df = pd.DataFrame(d)

    final = final.append(df, ignore_index = True)
```
